# Data perprocessing and cleaning

```python
df_encoding = pd.get_dummies(df, drop_first=True)
```

```python
df_encoding
```

```python
df_encoding = df_encoding.astype(int)
```

```python
df_encoding
```

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
sc = StandardScaler()
df_encoding[numerical_cols] = sc.fit_transform(df_encoding[numerical_cols])
df_encoding.head()
```

```python
df_encoding.columns
```

```python
X = df_encoding.drop('HeartDisease', axis=True)
y = df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
sc = StandardScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.fit_transform(X_test)
```

```python
models = {
          "lr":LogisticRegression(), 
          "nb":GaussianNB(), 
          "dt":DecisionTreeClassifier(), 
          "svm":SVC(probability=True), 
          "knn":KNeighborsClassifier()
          }
```

```python
results = []
```

```python
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, y_pred=y_pred)
    f1 = f1_score(y_test, y_pred=y_pred)
    results.append({
        'model':name,
        'Accuracy':round(acc,4),
        'F1_score':round(f1,4)
    })
```

```python
results
```

```python
import joblib
joblib.dump(models['knn'], 'KNN_heart.pkl')
joblib.dump(sc, 'scaler.pkl')
joblib.dump(X.columns.tolist(), 'columns.pkl')
```
